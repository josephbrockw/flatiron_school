### Let's do Regression 

Right here is the real deal. Let's perform a number of simple regression experiments between the chosen independent variables and the dependent variable (price). We shall do this is a loop and in every iteration, we shall pick one of the independent variables  perform following steps:

* Run a simple OLS regression between independent and dependent variables
* Plot a regression line on the scatter plots
* Plot the residuals using `sm.graphics.plot_regress_exog()`.
* Plot a Q-Q plot for regression residuals normality test 
* Store following values in array for each iteration:
    * Independent Variable
    * r_squared'
    * intercept'
    * 'slope'
    * 'p-value'
    * 'normality (JB)' 
* Comment on each output 

```python
print(smf.fit())
```

```python
# Your code here
results = [['iv', 'r_squared', 'intercept', 'slope', 'p_value', 'normality']]

for col in df:
    print('Boston Housing DataSet - Regression Analysis and Diagnostics for formula: medv~',col)
    print('-------------------------------------------------------------------------------------')
    # Scatter plot of data with regression line
    # scatter plot
    plt.scatter(df[col], df['medv'], c='b')
    f = 'medv~' + col
    model = smf.ols(formula=f, data=df).fit()
    col_min_max = pd.DataFrame({col: [df[col].min(), df[col].max()]})
    prediction = model.predict(col_min_max)
    # regression line
    plt.plot(col_min_max, prediction, c='r')
    # show scatter plot with regression line
    plt.show()
    
    # Plot the risiduals
    fig = plt.figure(figsize=(15,8))
    fig = sm.graphics.plot_regress_exog(model, col, fig=fig)
    plt.show()
    
    # Plot a Q-Q plot for regression residuals normality test
    resid = model.resid
    fig = sm.graphics.qqplot(resid, dist=stats.norm, line='45', fit=True,   )
    fig.show()
    
    # store values
    results.append([col, model.rsquared, model.params[0], model.params[1], model.pvalues[1], sms.jarque_bera(model.resid)[0]])
    ## independent variable
    #iv.append()
    ## r_squared
    ## intercept
    ## slope
    ## p-value
    ## normalit (JB)
    
```
