```python
import numpy as np
import random
import statsmodels.api as sm
import rpy2.robjects as robjects
import matplotlib.pyplot as plt
random.seed(5)
```

We'll store a list of errors. A list of numbers that we'd expect from subtracting our actual data from our predicted value (y_hat

```python
X = []
for i in range(1000):
    X.append(random.uniform(0.1, 5))

e = np.random.normal(loc=0, scale=0.2, size=1000) # loc = mean, scale = standard deviation, size = n
```

When using real data, the process of log transforming a Series is pretty straight forward. If I have my dependent variable saved in a variable `y`, I'd run the following code:

`y_log = np.log(y)`

We're lucky to have a libray like numpy that makes the process that simple. Since we're creating our own mock data, we'll also have to create a mock dependent variable. To simulate a log transformed dependent variable, we'll create a list of normally distributed data. The advantage of this for demonstration purposes is that we can get data based on a model that we define. That way, when we run our code, we know that model that the library **statsmodels** builds for us is correct.

```python
y = list(map(lambda x,e: 1.2 + 0.2 * x + e, X, e))
```

```python
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()

model.summary()
```

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, 1, fig=fig)
plt.show()
```

```python
sq_error = list(map(lambda x: x**2, model.resid))
rse = np.sqrt(np.sum(sq_error) / int(model.df_resid))
y_int = model.params[0]
slope = model.params[1]
print('y intercept: ', y_int)
print('Slope: ', slope)
print('Residual Standard Error: ',rse)
print('Degrees of Freedom: ', int(model.df_resid))
```
